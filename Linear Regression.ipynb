```python
#Data Preparation
```

```python
#Car Name
#Geting the Brand name
CarPrice['CarName']=CarPrice.CarName.apply(lambda strName: strName.split(" ")[0].strip())
```

```python
CarPrice['CarName'].value_counts()
```

```python
#Categorical column with 28 different values, grouping the brands in bins of mean value of price

carname=CarPrice[['CarName','price']]
carnamegrp=carname.groupby('CarName').mean().reset_index().sort_values(by="price",ascending=False)
CarNameByPriceMean=list(carnamegrp.CarName)
carnamegrp['MeanGrp']=carnamegrp['price']//10000
```

```python
carnamegrp
```

```python
#Method to grouping the car into bins
def CarNameCategory(carname):
    return "CarDiv"+str(int(carnamegrp[carnamegrp["CarName"]==carname].MeanGrp))
```

```python
CarPrice['CarName']=CarPrice.CarName.apply(CarNameCategory)
```

```python
#Verifying
CarPrice['CarName'].value_counts()
```

```python
#Creating Dummy Varaiables for Car Mean Price
CarName = pd.get_dummies(CarPrice['CarName'], drop_first = True)
CarPrice = pd.concat([CarPrice, CarName], axis = 1)
CarPrice.drop(['CarName'], axis = 1, inplace = True)
```

```python
CarPrice.fuelsystem.value_counts()
```

```python
#Creating Dummy Varaiables for fuelsystem
fuelsystem = pd.get_dummies(CarPrice['fuelsystem'], drop_first = True)
CarPrice = pd.concat([CarPrice, fuelsystem], axis = 1)
CarPrice.drop(['fuelsystem'], axis = 1, inplace = True)
```

```python
CarPrice.enginetype.value_counts()
```

```python
#Creating Dummy Varaiables for enginetype
enginetype = pd.get_dummies(CarPrice['enginetype'], drop_first = True)
CarPrice = pd.concat([CarPrice, enginetype], axis = 1)
CarPrice.drop(['enginetype'], axis = 1, inplace = True)
```

```python
CarPrice.fueltype.value_counts()
```

```python
#Creating Dummy Varaiables for fueltype
fueltype = pd.get_dummies(CarPrice['fueltype'], drop_first = True)
CarPrice = pd.concat([CarPrice, fueltype], axis = 1)
CarPrice.drop(['fueltype'], axis = 1, inplace = True)
```

```python
CarPrice.aspiration.value_counts()
```

```python
#Creating Dummy Varaiables for aspiration
aspiration = pd.get_dummies(CarPrice['aspiration'], drop_first = True)
CarPrice = pd.concat([CarPrice, aspiration], axis = 1)
CarPrice.drop(['aspiration'], axis = 1, inplace = True)
```

```python
CarPrice.carbody.value_counts()
```

```python
#Creating Dummy Varaiables for carbody
carbody = pd.get_dummies(CarPrice['carbody'], drop_first = True)
CarPrice = pd.concat([CarPrice, carbody], axis = 1)
CarPrice.drop(['carbody'], axis = 1, inplace = True)
```

```python
CarPrice.enginelocation.value_counts()
```

```python
#Creating Dummy Varaiables for enginelocation
enginelocation = pd.get_dummies(CarPrice['enginelocation'], drop_first = True)
CarPrice = pd.concat([CarPrice, enginelocation], axis = 1)
CarPrice.drop(['enginelocation'], axis = 1, inplace = True)
```

```python
#Method to Convert numbers in string to numeric
def numStr(str):
    str=str.lower()
    if str=="one":
        return 1
    elif str=="two":
        return 2
    elif str=="three":
        return 3
    elif str=="four":
        return 4
    elif str=="five":
        return 5
    elif str=="six":
        return 6
    elif str=="seven":
        return 7
    elif str=="eight":
        return 8
    elif str=="nine":
        return 9
    elif str=="ten":
        return 10
    elif str=="eleven":
        return 11
    elif str=="twelve":
        return 12
    else:
        return 0
    
```

```python
CarPrice.doornumber.value_counts()
```

```python
#Convert from string Numbers to numeric
CarPrice['doornumber']=CarPrice.doornumber.apply(numStr)
```

```python
CarPrice.cylindernumber.value_counts()
```

```python
#Convert from string Numbers to numeric
CarPrice['cylindernumber']=CarPrice.cylindernumber.apply(numStr)
```
